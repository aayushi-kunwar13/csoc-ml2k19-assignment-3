```python
#CLASSIFICATION PROBLEM
```

```python
wine = pd.merge(data1,data2,how='outer',on=None)
wine
```

```python
#wine.hist()
```

```python
x = wine.drop('quality',axis =1)
```

```python
x
```

```python
y= wine['quality']
```

```python
y
```

```python
#create a validation dataset
```

```python
x_train , x_test , y_train , y_test =train_test_split(x, y, test_size = 0.2, random_state = 7)
```

```python
models=[]
models.append(('LR',LogisticRegression()))
models.append(('KNN',KNeighborsClassifier()))
models.append(('CART',DecisionTreeClassifier()))
models.append(('SVM',SVC()))
models.append(('NB',GaussianNB()))
```

```python
results =[]
names =[]
for name,model in models:
    kfold = KFold(n_splits =10,random_state =7)
    cv_results = cross_val_score(model, x_train , y_train ,cv=kfold , scoring = 'accuracy')
    results.append(cv_results)
    names.append(name)
    msg ="%s: %f (%f)"%(name ,cv_results.mean() , cv_results.std())
    print(msg)
```

```python
#Since the accuracy of DECISION TREE CLASSIFIER is highest so using the DECISION TREE CLASSIFIER
```

```python
#Make the Prediction
```

```python
cart = DecisionTreeClassifier()
cart.fit(x_train , y_train)
```

```python
predictions = cart.predict(x_test)
```

```python
print(accuracy_score(y_test , predictions))
#print(confusion_matrix(y_test , predictions))
#print(classification_report(y_test , predictions))
```

```python
# We can see that it looks like DECISION TREE CLASSIFIER has the largest estimated accuracy score
# ﬁnal accuracy score
```

```python
#REGRESSION
```

```python
from sklearn import metrics        
```

```python
x_train, x_test ,y_train,y_test = train_test_split(x ,y ,test_size = 0.2,random_state =7 )
```

```python
models=[]
models.append(('LR',LinearRegression()))
models.append(('RD',Ridge()))
models.append(('LS',Lasso()))
models.append(('SVM',SVC()))
models.append(('EN',ElasticNet()))
models.append(('CART',DecisionTreeClassifier()))
models.append(('KNN',KNeighborsClassifier()))
```

```python
results =[]
names =[]
for name,model in models:
    model.fit(x_train , y_train)
    predictions = model.predict(x_test)
    d = metrics.mean_squared_error(y_test, predictions)
    msg ="%s: %f "%(name , d)
    print(msg)
    
```

```python
# loss IN LINEAR REGRESSION is minimum 
```

```python
#answer:
#min loss = LR: 0.518327 
```
